```python
import json

with open("labelledData.json","r") as inputfile2:
    outputDicts = json.load(inputfile2)
inputfile2.close()
```

```python
with open("labelledTestData.json","r") as inputfile2:
    outputDictstest = json.load(inputfile2)
inputfile2.close()
```

```python
with open("labelledValidationData.json","r") as inputfile2:
    outputDictsVal = json.load(inputfile2)
inputfile2.close()
```

```python
test = []
a = 0
b = 0
c = 0 
d = 0
skip = 0
valdata = []

for i in outputDictsVal:
    if i["review"] in test:
        continue
    skip = 0
    for j in range(len(outputDictstest)): # check if the review is in the test data
        if i["review"] == outputDictstest[j]["review"]:
            skip = 1
        continue	
    if skip == 1: # if the review is in the test data, skip it
        continue
    
    test.append(i["review"])
    if (i["polarity"] =='Neutral'):
        i["polarity"] = 0
        a +=1
    if (i["polarity"] =='Positive' ):
    
        i["polarity"] = 1
        b +=1
    if (i["polarity"] =='Negative' ):

        i["polarity"] = 2
        c +=1
    if (i["polarity"] =='Mixed' ):
        
        i["polarity"] = 3
        d +=1
    
    valdata.append(i)
    

print(b,a,d,c)
print(len(valdata))
with open("mixed_val.json","w") as outputfile:
    jsonOutput = json.dumps(valdata,indent=4)
    outputfile.write(jsonOutput)
```
